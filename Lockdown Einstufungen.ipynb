## Ermittlung der Corona Maßnahmen und Einteilung der Stärke des Lockdowns

Wann wird von einem von welchen Lockdown gesprochen?
Ermittlung der Kriterien, die die stärke des Lockdowns bestimmen. Historie der Regelungen werden aus den Verordungen/Gesetzen etc ermittelt, die unter https://de.wikipedia.org/wiki/Liste_der_infolge_der_COVID-19-Pandemie_erlassenen_deutschen_Gesetze_und_Verordnungen#Baden-W%C3%BCrttemberg zur verfügung stehen. Auch anhand von https://de.wikipedia.org/wiki/Corona-Verordnung_(Baden-W%C3%BCrttemberg) (Wichtigste Punkte mit Datum aufgeführt, verlinkt die Gesetzestexte)
Ermittlung des Zeitraumes, in dem die Kriterien der Verornungen gelten.

### Was wurde in den Verordnungen bestimmt?

**_Von 2020/03/17 bis 2020/04/19_** (Verordnung tritt 2020/06/15 ausser Kraft)
- Schulen geschlossen (weitere Betreuung für Kinder von Eltern in Kritischen Sektoren)
- Hochschulen geschlossen (einschließlich Mensa/Cafeteria und Landes-/Hochschulbibliotheken)
- Verbot von Versammlungen / Veranstaltungen (sportm/Music/VHS/Reisen/Ausserschulischer Bereich/Religiöse Versammlungen)
- Schliessung von Einrichtungen (Kultur / Bildungseinrichtungen / Kinos / Schwimmbäder / Sportanlagen / Jugendhäuser / Bibliotheken / Casinos / Bordelle / Clubs / Kneipen / Messen / Einzelhandel / Spielplätze)
- Einschränkungen Gaststätten: Grundsätzlich Untersagt. (ausser wenn 1,5m zwischen Tischen und zwischen Gästen. Zeitbeschränk 6-18H)
- Schutzmaßnamen gefärdete Personen: Keine Besuche
- Betretungsverbot jeglicher (öffentlicher) Einrichtungen für Personen, die inerhalb von 14 Tagen in einem Risikogebiet waren
- Versammlungen über 100 PErsonen

**_Änderungen der Masnahmen am 2020/03/21_**
- Verweilen von mehr als 3 Personen im Öffentlichem Raum untersagt.
- Private versammlungen mit max 5 PErsonen
- Keine Fahrten aus Risikogebieten nach BW, ausser für Berufszwecke, Lieferungen oder per. Gründe (Todesfälle)
- Schliessung der Gaststätten (Abholungen/Lieferung erlaubt)
- Schliessung Frisöre und Ähnliches
- Schliessung Hotel für Touristen (dienstliche Aufenthalte ok)
- Wiedereröffnung ausgabestätten TAfel

**_Änderungen vom 2020/03/22_**
- Aufenthalt im öffentlichem Raum mit nur noch einer weiterem Person aus einem Anderen Haushalt
- Mindestabstand von 1,5m zu weiteren Personen im Öffentlichem Raum
- Wiedereröffnung Betriebskantienen

**_Änderungen am 2020/03/29_**
- Verstöße gegen Covid yMaßnahmen == Ordnungswiedrigkeit.
- Zustrittskontrolle und abstandsregelungen für Einzelhändler

**_Änderungen am 2020/04/10_**
- Einschränkung Zahnärztlicher Behandlungen (Nur Notfälle)

**_Änderungen am 2020/_**
- Wiederaufnahme des Studiensbetriebs ab 2020/04/20 in digitaler Form
- Verlängerung der Schließung der Schulen bis 2020/05/03

**_Von 2020/05/11 bis 2020/06/15_**
- Schulen geschlossen, ausser wenn golgende Maßnahmen eingehalten werden können: (bis 2020/06/15)
    - Vorbereitung auf Prüfungen, wenn 1,5m zwischen Personen mögich sind
    - Zeitliche Staffelung der Pausen
    - Möglichkeiten zum Händewaschen/Desinfizieren -- Lüften
    - Tägliche Reinigung
- Kein Betrieb für Kleinere Kinder, ausser Notbetreuung bei Berufstätigen Eltern in bestimmten Sektoren
- Keine Präsentzvorlesungen bis 2020/05/24, Prüfungen unter Hygiene VOrschrifften erlaubt
- Auffenthalt im Öffentlichem Raum (bis 2020/06/05): 
    - 1,5 m Abstand
    - nur eiigener Haushalt + 1 anderer
- Maskenpflicht ab 6 Jahren (bis 2020/06/05)
    - In ÖPVN, Bus-/Bahnsteigen, Flughäfen
    - Einkaufsläden
- Ausserhalb des Öffentlichen Raums (bis 2020/06/05):
    - Verboten für mehr als 5 Personen
    - Ausgenommen Direkte Verwandte, eigener Haushalt
- Kirchen etc. offen
- Einschränkungen betrieb Einrichtungen (bis 2020/05/24):
    - Theater
    - Bildungseinrichtungen
    - Kinos
    - Schwimmbad
    - Sportanlagen
    - Jugendhäuser
    - Bordelle
    - Gaststätten / Bars
    - Messen
    - Beherbergungsbetriebe (ausser geschäftliche Beherbergung
    - Reisebusse
- Ausgenommen davon:
    - AUsser-Haus Verkauf von Gastsätten
    - Abhol-/Lieferservice
    - Kantinen in Betrieben
    - Speisewirtschaften (ab 2020/05/18)
    - Museen
    - Bibliotheken
    - Autokinos
    - Zoos
    - Öffentliche Spielplätze
    - Häfen / fligplätze
    - Freiluftsportanlagen
    - Freitzeiteinrihtungen (ab 2020/05/18)
    - Campingplätze (ab 2020/05/18)
